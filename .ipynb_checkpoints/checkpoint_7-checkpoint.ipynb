# Pr√©diction de la survie des passagers du titanic

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
data = pd.read_csv(r'C:\Users\IDEAPAD\Desktop\GOMYCODE\SML\Titanic Forest\Resources\titanic-passengers.csv')
data.head()
```

```python
data.describe()
```

```python
data = data.drop(['PassengerId', 'Name', 'Cabin', 'Ticket', 'Embarked'], axis = 1)#Suppression des colonnes jug√©es inutiles
data = data.dropna(axis=0) # Suppression des ligne contenant des valeurs nulles
data.isnull().sum()
```

```python
data.shape
```

```python
data.head()
```

```python
#Transfom Cat data to Num data
encoder = LabelEncoder()
data['Survived'] = encoder.fit_transform(data['Survived']) # Yes : 1 , No : 0
data.head()
```

```python
#Corr√©lation entre l'age et le sexe dans la survie de l'individu
sns.set_theme(style="whitegrid")

survived = 'survived'
not_survived = 'not survived'
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10, 4))
women = data[data['Sex']== 'female'] #DF contenant seulement les femmes 
men = data[data['Sex']== 'male'] #DF contenant seulement les hommmes 

ax = sns.histplot(women[women['Survived']==1].Age, bins=18, label = survived, ax = axes[0], kde =False, color='Green')
ax = sns.histplot(women[women['Survived']==0].Age, bins=40, label = not_survived, ax = axes[0], kde =False, color='Red')
ax.legend()
ax.set_title('Femme')

ax = sns.histplot(men[men['Survived']==1].Age, bins=18, label = survived, ax = axes[1], kde = False, color='Green')
ax = sns.histplot(men[men['Survived']==0].Age, bins=40, label = not_survived, ax = axes[1], kde = False, color='Red')
ax.legend()
ax.set_title('Homme')
```

```python
#Importance de la classe dans la survie de l'individu
sns.barplot(x='Pclass', y='Survived', data=data)
```

```python
sns.histplot(data, x="Sex") # Plus d'homme que de femmes
```

```python
sns.histplot(data, x="Pclass", hue='Sex', multiple="dodge") #Plus d'hommes dans la 3√®me classe qui a compt√© le moins de survivant
```

```python
#Conna√Ætre le nombre d'enfant
def titanic_children(passenger):
    age , sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex

data['Person'] = data[['Age', 'Sex']].apply(titanic_children,axis=1)
data.tail()
```

```python
sns.histplot(data, x="Pclass", hue='Person', multiple="dodge")
```

```python
#V√©rifier qu'un passager √† de la famille ou pas
#SibSp : Nombre de Fr√®res et Soeurs et Parch : Nombre de Parents √† bord du Titanic
data['Alone'] = data['SibSp'] + data['Parch']
data['Alone'].loc[data['Alone'] > 0] = 'Has a family'
data['Alone'].loc[data['Alone'] == 0] = 'Has no family'
data.tail()
```

```python
#Conversion des Data Cat√©gorielles en Data Num√©riques
data['Sex'] = encoder.fit_transform(data['Sex']) # Male : 1 , Female : 0
data['Person'] = encoder.fit_transform(data['Person']) # Male :2 , Female : 1 , Child : 0
data['Alone'] = encoder.fit_transform(data['Alone']) # Has no Family : 1, has Family : 0
```

```python
#S√©paration de nos donn√©es
X = data.drop(['Survived'],axis=1)
Y = data['Survived']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=40)
```

# Arbre de D√©cision

```python
model_tree = DecisionTreeClassifier()
model_tree.fit(x_train, y_train)
y_pred = model_tree.predict(x_test)
print((" Accuracy : {} %").format(accuracy_score(y_test, y_pred) * 100))
```

```python
#Trac√© de l'arbre
fig = plt.figure(figsize=(25, 20))
_ = tree.plot_tree(model_tree, filled=True)
```

```python
#Modification des param√®tres de l'arbre
model_tree = DecisionTreeClassifier(criterion='gini', splitter='random', max_leaf_nodes=10, min_samples_leaf=5, max_depth=5)
model_tree.fit(x_train, y_train)
y_pred = model_tree.predict(x_test)
print(("Accuracy : {} % ").format(accuracy_score(y_test, y_pred)*100))
```

```python
#Trac√© de l'arbre
fig = plt.figure(figsize=(25, 20))
_ = tree.plot_tree(model_tree, feature_names=['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Person', 'Alone'] ,filled=True)
```

# Random Forest

```python
clf = RandomForestClassifier(n_estimators=10)
clf.fit(x_train, y_train)
pred = clf.predict(x_test)

confusion_matrix = pd.crosstab(y_test, pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

```python
print(classification_report(y_test, pred)) # Accuracy = 87%
```
